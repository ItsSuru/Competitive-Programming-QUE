# Friends Meeting

https://codeforces.com/contest/931/problem/A

```python
n=int(input())
m=int(input())
if m-n==1 or n-m==1:
    print(1)
else:
    x=abs(n-m)
    if x%2==0:
        x=x//2
        print((x*x)+x)
    else:
        x=(x//2)+1
        print(x*x)
```
