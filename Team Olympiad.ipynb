# Team Olympiad

https://codeforces.com/contest/490/problem/A

```python
n = int(input())
lst = list(map(int,input().split()))
c1 = lst.count(1)
c2 = lst.count(2)
c3 = lst.count(3)
l1,l2,l3=[],[],[]
teams = min(c1,c2,c3)
print(teams)
for i in range(len(lst)):
	if lst[i]==1:
		l1.append(i+1)
	elif lst[i]==2:
		l2.append(i+1)
	else:
		l3.append(i+1)
for j in range(teams):
	print(l1[j],l2[j],l3[j])
```
