# Fingerprints

https://codeforces.com/problemset/problem/994/A

```python
n,m=map(int,input().split())
l=list(map(int,input().split()))
k=list(map(int,input().split()))
p=[]
u=[]
for i in k:
    for j in l:
        if i==j:
            p.append(l.index(i)+1)
        else:
            pass
        
p.sort()        
s=[]
 
for q in p:
    q-=1
    s.append(q)
    
for b in s:
    print(l[b],end=" ")
```
