# Beautiful String

https://codeforces.com/problemset/problem/1265/A

```python
for u in range(int(input())):
    s = input()
    s=list(s)
    p=''
    n=''
    for i in range(len(s)):
        if s[i]=='?':
            if i>0:
                p=s[i-1]
            if i<len(s)-1:
                n=s[i+1]
            for j in ['a','b','c']:
                if j!=p and j!=n:
                    s[i]=j
                    break;
    f=0
    for i in range(len(s) - 1):
        if (s[i] == s[i+1]):
            print("-1")
            f=1
            break
    if (f==0):  
        print("".join(s))  
            
```
