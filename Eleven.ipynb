# Eleven

https://codeforces.com/problemset/problem/918/A

```python
n=int(input())
p=0
l=1
k=[]
j=[]
o=[]
if n==0:
    print(n)
elif n==1:
    print('O')
elif n==2:
    print('OO')
elif n==3:
    print('OOO')    
else:
    if (n==0):
        print(n)
        exit(0)
    
    else:
        for i in range(n+1):
            ans=p+l
            if p==0:
                pass
            else:
                k.append(p)       
            p=l
            l=ans
        
    for i in k:
        if i<=n:
            j.append(i)
        else:
            pass
 
    for i in range(1,n+1):
        o.append('o')
    
    for u in j:
        o[u-1]="O" 
        
    for b in o:
        print(b,end='')
```
