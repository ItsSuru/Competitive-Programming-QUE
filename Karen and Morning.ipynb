# Karen and Morning

https://codeforces.com/problemset/problem/816/A

```python
n,k = map(int, input().split(":"))
ans = 0
while (True):
    rk = int(str(k)[::-1])
    if(len(str(k)) == 1):
        rk = int(str(k)+'0')
    if(rk == n):
        break
    elif(k < 59):
        k += 1
    elif(k == 59):
        k = 0
        if(n < 23):
            n += 1
        else:
            n = 0
    ans += 1
print(ans)
```
