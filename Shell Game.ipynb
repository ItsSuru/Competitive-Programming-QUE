# Shell Game

https://codeforces.com/problemset/problem/777/A

```python
n=int(input())
m=int(input())
summ=n%6
if summ==0:
    print(m)
elif summ==1:
    if m==1:
        print(0)
    elif m==0:
        print(1)
    else:
        print(2)
elif summ==2:
    if m==1:
        print(2)
    elif m==0:
        print(1)
    else:
        print(0)
elif summ==3:
    if m==1:
        print(1)
    elif m==0:
        print(2)
    else:
        print(0)
elif summ==4:
    if m==1:
        print(0)
    elif m==0:
        print(2)
    else:
        print(1)
elif summ==5:
    if m==1:
        print(2)
    elif m==0:
        print(0)
    else:
        print(1)
else:
    pass
```
