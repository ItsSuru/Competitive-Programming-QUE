# Rounding

https://codeforces.com/problemset/problem/898/A

```python
n=int(input())
p=str(n)
l=p[len(p)-1]
l=int(l)
if l==0:
    print(n)
else:
    if l<=5:
        n=n-l
        print(n)
    else:
        l=10-l
        n=n+l
        print(n)
```
