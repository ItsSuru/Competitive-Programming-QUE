# Diversity

https://codeforces.com/problemset/problem/844/A

```python
s=input()
n=int(input())
k=[]
alp=[]
p=0
for i in s:
    k.append(i)
for j in range(97,123):
    alp.append(chr(j))
for e in alp:
    if e in k:
        p+=1
if n>len(s):
    print("impossible")
elif n<p:
    print(0)
else:
    print(n-p)
```
