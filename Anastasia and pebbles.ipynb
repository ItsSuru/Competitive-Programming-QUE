# Anastasia and pebbles

https://codeforces.com/problemset/problem/789/A

```python
n,m=map(int,input().split())
k=list(map(int,input().split()))
j=0
for i in k:
    if i<=m:
        j+=1
    else:
        if i>m:
            if i%m==0:
                l=i//m
                j+=l
            else:
                l=i//m+1
                j+=l             
if j%2==0:
    b=j//2
    print(b)
else:
    b=j//2+1
    print(b)
```
