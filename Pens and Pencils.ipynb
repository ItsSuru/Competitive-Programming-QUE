# Pens and Pencils

https://codeforces.com/problemset/problem/1244/A

```python
for i in range(int(input())):
    a,b,c,d,k=map(int,input().split())
    y=d
    u=c
    o=1
    p=1
 
    while d<=b:
        d=y
        d=d*p
        if d>=b:
            break
        else:
            p+=1
       
    while c<=a:
        c=u
        c=c*o
        if c>=a:
            break
        else:
            o+=1        
        
    if p+o<=k:
        print(o,p)
    else:
        print("-1")
```
