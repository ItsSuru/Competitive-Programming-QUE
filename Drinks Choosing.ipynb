# Drinks Choosing

https://codeforces.com/problemset/problem/1195/A

```python
n,m=map(int,input().split())
k=[]
p=[]
t=[]
b=[]
for i in range(n):
    c=int(input())
    k.append(c)
 
for j in range(1,max(k)+1):
    if j in k:
        p.append(k.count(j))
    else:
        p.append(0)        
 
for i in p:
    if i%2==0:
        t.append(i)
    elif i>=3:
        if i%2==0:
            t.append(i)
        else:
            t.append(i-1)
            b.append(1)
    else:
        b.append(i)
 
if sum(b)%2==0:
    b=sum(b)//2
else:
    b=(sum(b)//2)+1
    
print(sum(t)+b)
```
