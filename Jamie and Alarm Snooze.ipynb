# Jamie and Alarm Snooze

https://codeforces.com/problemset/problem/916/A

```python
num = int(input())
h, m = map(int, input().split())
 
ans = 0
while "7" not in str(h)+str(m):
    if m < num:
        m = 60+m-num
        ans += 1
        if h == 0:
            h = 23
        else:
            h -= 1
    else:
        m -= num
        ans += 1
print(ans)
```
