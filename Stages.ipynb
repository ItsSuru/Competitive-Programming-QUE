# Stages

https://codeforces.com/contest/1011/problem/A

```python
m,k=map(int,input().split())
n=list(input())
n.sort
p=[]
h=[]
v=[]
# getting no of str
for i in range(97,123):
    if chr(i) in n:
        p.append(i) 
#appending only required no        
o=min(p) 
h.append(o)
u=0
for _ in p:
    if _ > o+1:
        if u<(k-1):
            h.append(_)
            o=_
            u+=1
#if length of h == k then printing total else -1          
if len(h)==k:
    for e in h:
        v.append(e-96)
    g=0    
    for d in v:
        g=d+g
    print(g)
else:
    print(-1)
```
