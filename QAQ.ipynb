# QAQ

https://codeforces.com/contest/894/problem/A

```python
s =  input()
ans=0
for i in range(len(s)):
    if s[i]=="Q":
        for j in range(i,len(s)):
            if s[j]=="A":
                for k in range(j,len(s)):
                    if s[k]=="Q":
                        ans+=1
print(ans)
```
