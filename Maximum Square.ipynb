# Maximum Square

https://codeforces.com/problemset/problem/1243/A

```python
for i in range(int(input())):
    k=int(input())
    l=list(map(int,input().split()))
    p=0
    for i in l:
        if l[0]==i:
            p+=1
    if p==len(l):
        print(l[0])
    else:
        j=[]
        for u in range(1,max(l)+1):
            o=0
            for i in l:
 
                if u<=i:
                    o+=1
            if o >=u:
                j.append(u)
        
        print(max(j))  #pypy
```
