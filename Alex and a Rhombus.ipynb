# Alex and a Rhombus

https://codeforces.com/problemset/problem/1180/A

```python
n=int(input())
p=[]
if n==1:
    print(1)
else: 
    m=n-1
    for i in range(1,m+1):
        k=4*i
        p.append(k)
    print(sum(p)+1)
```
