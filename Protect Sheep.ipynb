# Protect Sheep

https://codeforces.com/contest/948/problem/A

```python
r,c=map(int,input().split())
l=[]
for i in range(r):
    q=input()
    if "SW"in q or "WS"in q:exit(print("No"))
    q=q.replace(".","D")
    l+=[q]
b=['']*c
for i in range(c):
    for j in range(r):
        b[i]+=l[j][i]
    if "SW" in b[i] or "WS" in b[i]:
        exit(print("NO"))
print("YES")
for i in range(r):
    print(l[i])
```
