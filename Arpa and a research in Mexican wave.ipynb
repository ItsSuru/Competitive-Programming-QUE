# Arpa and a research in Mexican wave

https://codeforces.com/contest/851/problem/A

```python
n,m,k=map(int,input().split())
if k<=n and k<=m:
    print(k)
elif k<=n and k>=m:
    print(m)
elif k>n:
    k=k-n
    print(m-k)
```
