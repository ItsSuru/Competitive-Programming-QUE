# Salem and Sticks

https://codeforces.com/problemset/problem/1105/A

```python
a = int(input())
sticks= list(map(int, input().split()))
minn = []
for t in range(1, 101):
    minn.append(0)
    for a in sticks:
        p = abs(a - t)
        if p != 0:
            p -= 1
        minn[t - 1] += p
print(minn.index(min(minn)) + 1, min(minn))
```
