#  Nastya Is Reading a Book

https://codeforces.com/problemset/problem/1136/A

```python
n= int(input())
l=[]
p=[]
for i in range(n):
    m,k=list(map(int,input().split()))
    l.append(m)
    l.append(k)
    p.append(k)
    
j=int(input())
 
for u in range(len(p)):
    if j<=p[u]: 
        g=p.index(p[u])
        o=p[g:]
        print(len(o))
        break
    else:
        pass
```
