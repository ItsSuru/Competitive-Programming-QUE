# The Rank

https://codeforces.com/problemset/problem/1017/A

```python
n=int(input())
j=[]
for i in range(n):
    a,b,c,d=list(map(int,input().split()))
    e=a+b+c+d
    j.append(e)
    
l=max(j)
if j[0]==l:
    print("1")
else:
    m=j[0]
    g=sorted(j,reverse=True)
    f=g.index(m)
    p=g.count(m)
    if p>=2:
        p-=1
        f+=1
        print(f)
    else:
        f+=1
        print(f)
```
