# Odds and Ends

https://codeforces.com/problemset/problem/849/A

```python
n=int(input())
l=list(map(int,input().split()))
if n%2==0:
    print("NO")
else:
    if l[0]%2!=0 and l[n-1]%2!=0:
        print("YES")
    else:
        print("NO")
```
