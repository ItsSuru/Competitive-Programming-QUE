```python
n=int(input())
m=input()
z=0
o=0
for i in m:
    if i=='0':
        z+=1
    else:
        o+=1
if len(m)==1:
    print(1)
    print(m[0])
elif (z>o) or (o>z):
    print(1)
    print(m)
else:
    print(2)
    print(m[0],m[1:])
```
