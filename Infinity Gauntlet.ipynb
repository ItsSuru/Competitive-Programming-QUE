# Infinity Gauntlet

https://codeforces.com/problemset/problem/987/A

```python
lst1=[]
n = int(input())
for i in range(0, n):
            ele = input()
            lst1.append(ele)
 
lst = ["red","purple","yellow","green","blue","orange"]
x=[]
i=0
for j in range(len(lst)):
    if lst[i] in lst1:
        pass
    else:
        x.append(lst[i])
    i+=1
print(len(lst)-len(lst1))        
for i in x:
    if i == "green":
        print("Time")
    if i ==  "purple":
        print("Power")
    if i ==  "blue":
        print("Space")
    if i ==  "orange":
        print("Soul")
    if i ==  "red":
        print("Reality")
    if i ==  "yellow":
        print("Mind") 
```
