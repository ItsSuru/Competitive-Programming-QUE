```python
n=int(input())
k=n*4
m=k+n
if (m%3==0) or (m%2==0):
    print(m,k)
else:   
    m+=1
    k+=1
    if (k%3==0) or (k%2==0):
        print(m,k)
    else:
        m+=2
        k+=2
        if (k%3==0) or (k%2==0):
            print(m,k)
        else:
            m+=1
            k+=1
            print(m,k)
```
