# Fair Division

https://codeforces.com/gym/323895/problem/B

```python
try:
    for i in range(int(input())):
        n=int(input())
        l=list(map(int,input().split()))
        if(sum(l)%2==0):
            a=len(set(l))
            if(a==1):
                if(len(l)%2==0):
                    print("yes")
                else:
                    print("no")
            else:
                print("yes")
        else:
            print("no")
except:
    print("wrong")
```
