# Ehab Fails to Be Thanos

https://codeforces.com/contest/1174/problem/A

```python
n=int(input())
m=list(map(int,input().split()))
l=[]
if min(m)==max(m):
    print(-1)
else:
    for i in range(n):
        p=max(m)
        l.append(p)
        m.remove(p)
    h=l+m
    for j in h:
        print(j,end=" ")
```
