# Even Odds

https://codeforces.com/contest/318/problem/A

```python
n, m = map(int, input().split())
odds = (n + 1) // 2
if m <= odds:
    print(int(m * 2) - 1)
else:
    print(int((m - odds) * 2))
```

```python
n, k = map(int, input().split())
odd=0
if n%2==0:
    odd=n//2
else:
    odd=(n+1)//2
 
if k<=odd:
    ans=k*2-1
elif k>odd:
    ans=(k-odd)*2
print(ans)
```
