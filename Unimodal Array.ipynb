# Unimodal Array

https://codeforces.com/problemset/problem/831/A

```python
n=int(input())
l=[int(i) for i in input().split()]
i=1 
c=0
while i<n and l[i]>l[i-1]:
    i+=1 
    c+=1 
while i<n and l[i]==l[i-1]:
    i+=1 
    c+=1 
while i<n and l[i]<l[i-1]:
    i+=1 
    c+=1 
print('YES' if c==n-1 else 'NO')
```
