# Feed the cat

https://codeforces.com/problemset/problem/955/A

```python
t,m = map(int,input().split())
h,d,c,n = map(int,input().split())
total1=(t*60)+m
b=(h//n)
if h%n>0:
    b+=1
b*=c
 
if t<20:
    t1=(20-t)-1
    m=60-m
    total=((t1*60)+(m))*d
    l=h+total
    o=l//n
    if l%n>0:
        o+=1
    p=o*c
    p*=0.8
else:
    p=b-(b*0.2)
print(min(p,b)) 
```
